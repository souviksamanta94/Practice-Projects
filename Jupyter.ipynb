```python
import pandas as pd
import numpy as np
import AdvancedAnalytics
from AdvancedAnalytics import ReplaceImputeEncode

df = pd.read_excel("sonar3by5.xlsx")
print("First 15 data values : \n",df[0:15])
```

```python
data_map={}
# Generate the dictionary keys for the 60 sonar frequencies 'R1', 'R2', ...
for i in range(60):
    str = 'R'+"%i" %(i+1)
    data_map[str]=['I',(0,1)]
# Lastly add the key for the target. In this data this is labeled 'object'
data_map['object'] = ['B',('M', 'R'), [0,0]]
#Imputing data
rie = ReplaceImputeEncode(data_map=data_map,display=True)
df.rie = rie.fit_transform(df)          
```

```python
print("Imputed Data: \n", df.rie)
print("First 15 imputed data: \n",df.rie[1:15])  
```

```python
#Regression
from AdvancedAnalytics import logreg
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

y = np.asarray(df.rie['object'])
x = np.asarray(df.rie.drop('object',axis = 1))
#Creating validation set
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state=1234)
lr_train = LogisticRegression()
lr_train.fit(x_train,y_train)
print("\nTraining Data\nRandom Selection of 70% of Original Data")
logreg.display_binary_split_metrics(lr_train, x_train, y_train,x_test, y_test)
```

```python
y_hat = lr_train.predict(x_test)
print(" First 15 Predicted Values :", y_hat[0:15])
```
